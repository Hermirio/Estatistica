#### 2. Qual a diferença entre uma lista, um dicionário, um array numpy e uma series pandas?

##### Lista
na lista o operador *in* usa algoritmos diferentes para listas. Para listas, ele procura os elementos da lista em ordem.
##### Dicionario
Para dicionários, o Python usa um algoritmo chamado hashtable (tabela de dispersão), que tem uma propriedade notável: o operador in leva praticamente o mesmo tempo na busca, não importa quantos itens estejam no dicionário

##### Serie
Um serie e um array de uma dimensão, ou pode consederar tambem um coluno da tabela.

##### Array
Um Array e uma matriz de Valores de um mesmo tipo

#### (b) Crie uma função genérica para trocar duas colunas, sem codificar os nomes das colunas.

```python
def changeColumns( data, colA, colB ):
    valorA = data[colB].copy()
    data[colB] = data[colA]
    data[colA] = valorA

changeColumns( df, 'a', 'b')
df
```

#### 9. Obtenha as duas últimas linhas de df cuja soma da linha é maior que 100.

```python
df = pd.DataFrame(np.random.randint(10, 40, 60).reshape(-1, 4))
print(df)

def test(df):
    lista = []
    
    for i in range(len(df)):
        if ( df.iloc[-i].sum() > 100 ):
            lista.append(df.iloc[-i])

            if len(lista) == 2: 
                return lista

lista = test(df)
print(f"\nindice {lista[0].name} : {lista[0].sum()}\nindice {lista[1].name} : {lista[1].sum()}")
```
